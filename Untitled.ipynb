```python
#import necessary packages
import netCDF4
import xarray as xr
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import cartopy.crs as ccrs
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import imageio
import os
#Finish Importation of Packages

#Clear up error in MatPlotLib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
#Set Desired Range of years
min_year=2010
max_year=2010
#Derive numyears variable
numyears=max_year-min_year+1
```

```python
#Merge Datsets
Vwindfilepath='WindVelocities/Vwind/vwnd.10m.gauss.'
DSV=['']*numyears
Uwindfilepath='WindVelocities/Uwind/uwnd.10m.gauss.'
DSU=['']*numyears
fileext='.nc'
year=['']*numyears
for x in range(numyears):
    year[x]=str(min_year+x)
    
    DSV[x]=xr.open_dataset(Vwindfilepath+year[x]+fileext)
        
        
    DSU[x]=xr.open_dataset(Uwindfilepath+year[x]+fileext)
    
    if x==0:
        FullWind=FullWind=xr.merge([DSV[x],DSU[x]])
    else:
        FullWind=xr.merge([FullWind,DSV[x],DSU[x]])
        
    DSV[x].close()
    DSU[x].close()
    
#End Merging of Datasets
```

```python
#Add Variables to Wind Dataset
#WindMag gives the total wind magnitude
#WMI gives the number of days below 6 m/s
FullWind=FullWind.assign(WindMag = lambda FullWind: (FullWind.vwnd**2+FullWind.uwnd**2)**0.5)
FullWind=FullWind.assign(WMI = lambda FullWind: FullWind.vwnd*0)

#Creating test dataset
FullWind_test=xr.merge([FullWind.vwnd[0:10,0:10,0:10],FullWind.uwnd[0:10,0:10,0:10],FullWind.WindMag[0:10,0:10,0:10],FullWind.WMI[0:10,:,:]])



#---------------------------------------------------------------------
#Averaging Longitudinal data & Restricting Latitudes to Arctic Regions
#---------------------------------------------------------------------

#Pull dimensions for arrays
londims=int((len(FullWind.lon)/3))
latdims=int(len(FullWind.lat))
timedims=int(len(FullWind.time))

#Create arrays for averaged variables
NWM=np.zeros((timedims,latdims,londims))
NVW=np.zeros((timedims,latdims,londims))
NUW=np.zeros((timedims,latdims,londims))
NWMI=np.zeros((timedims,latdims,londims))

#Pull Coordinates 
time=FullWind.time
lat=FullWind.lat
lon=[0]*londims

#Write array for new longitudes
for x in range(londims):
    lon[x]=(FullWind.lon[3*x]+FullWind.lon[3*x+1]+FullWind.lon[3*x+2])/3

#Average Variables for new latitudes
for x in range(londims):
    NWM[:,:,x]=(FullWind.WindMag[:,:,3*x]+FullWind.WindMag[:,:,3*x+1]+FullWind.WindMag[:,:,3*x+2])/3
    NVW[:,:,x]=(FullWind.vwnd[:,:,3*x]+FullWind.vwnd[:,:,3*x+1]+FullWind.vwnd[:,:,3*x+2])/3
    NUW[:,:,x]=(FullWind.uwnd[:,:,3*x]+FullWind.uwnd[:,:,3*x+1]+FullWind.uwnd[:,:,3*x+2])/3
    NWMI[:,:,x]=(FullWind.WMI[:,:,3*x]+FullWind.WMI[:,:,3*x+1]+FullWind.WMI[:,:,3*x+2])/3
    
#Create xarray Data Arrays for variables
NWM=xr.DataArray(NWM, coords=[time,lat,lon], dims=['time','lat','lon'])
NVW=xr.DataArray(NVW, coords=[time,lat,lon], dims=['time','lat','lon'])
NUW=xr.DataArray(NUW, coords=[time,lat,lon], dims=['time','lat','lon'])
NWMI=xr.DataArray(NWMI, coords=[time,lat,lon], dims=['time','lat','lon'])

#Write dictionaries for Writing Dataset
data_vars={
    'vwnd':NVW,
    'uwnd':NUW,
    'WindMag':NWM,
    'WMI':NWMI,
}
coords={
    'time':time,
    'lat':lat,
    'lon':lon,
}

#Rewriting Dataset w/ averaged longitudes
FullWind_lonavg=xr.Dataset(data_vars, coords)

#---------------------------
#End Averaging of longitudes
#---------------------------

#Restricting Latitude to the Arctic regions
FullWind_arctic=xr.merge([FullWind_lonavg.vwnd[:,0:12,:], FullWind_lonavg.uwnd[:,0:12,:], FullWind_lonavg.WindMag[:,0:12,:], FullWind_lonavg.WMI[:,0:12,:]])
```

```python
def latsmooth(Dataset,SmoothingFactor):

    londims=int((len(Dataset.lon)/SmoothingFactor))
    latdims=int(len(Dataset.lat))
    timedims=int(len(Dataset.time))

    NWM=np.zeros((timedims,latdims,londims))
    NVW=np.zeros((timedims,latdims,londims))
    NUW=np.zeros((timedims,latdims,londims))
    NWMI=np.zeros((timedims,latdims,londims))

    time=Dataset.time
    lat=Dataset.lat
    lon=[0]*londims

    for x in range(londims):
        lon[x]=(Dataset.lon[3*x]+Dataset.lon[3*x+1]+Dataset.lon[3*x+2])/3

    for x in range(londims):
        NWM[:,:,x]=(Dataset.WindMag[:,:,3*x]+Dataset.WindMag[:,:,3*x+1]+Dataset.WindMag[:,:,3*x+2])/3
        NVW[:,:,x]=(Dataset.vwnd[:,:,3*x]+Dataset.vwnd[:,:,3*x+1]+Dataset.vwnd[:,:,3*x+2])/3
        NUW[:,:,x]=(Dataset.uwnd[:,:,3*x]+Dataset.uwnd[:,:,3*x+1]+Dataset.uwnd[:,:,3*x+2])/3
        NWMI[:,:,x]=(Dataset.WMI[:,:,3*x]+Dataset.WMI[:,:,3*x+1]+Dataset.WMI[:,:,3*x+2])/3

    NWM=xr.DataArray(NWM, coords=[time,lat,lon], dims=['time','lat','lon'])
    NVW=xr.DataArray(NVW, coords=[time,lat,lon], dims=['time','lat','lon'])
    NUW=xr.DataArray(NUW, coords=[time,lat,lon], dims=['time','lat','lon'])
    NWMI=xr.DataArray(NWMI, coords=[time,lat,lon], dims=['time','lat','lon'])

    data_vars={
        'vwnd':NVW,
        'uwnd':NUW,
        'WindMag':NWM,
        'WMI':NWMI,
    }
    coords={
        'time':time,
        'lat':lat,
        'lon':lon,
    }

    FullWind=xr.Dataset(data_vars, coords)

    return FullWind
```
